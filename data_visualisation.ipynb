```python
# importing relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
data = pd.read_csv("Recidivism__Beginning_2008.csv")
data.head(20)
```

```python
data.tail ()
```

```python
data.shape
```

```python
sns.histplot(data['Age at Release'],bins=30)
plt.title("Age distribution of prisoners at release")
plt.show()
```

```python
sns.boxplot(x='Age at Release', y='Gender', data=data)
plt.title("Age at Release by Gender")
plt.show()
```
